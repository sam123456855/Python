```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
def getSource(url):
    r = requests.get(url, verify=False)
    if r.status_code == 200:
        html = soup(r.text,'html.parser')
        return html
    else:
        return "unable to download"

getSource("https://infatica.io/blog/web-crawling-vs-web-scraping/")
```

```python
def getImages(html):
    images = html.find_all('img')
    if images:
        return images
    else:
        return []
```

```python
def main():
    url = 'https://infatica.io/blog/web-crawling-vs-web-scraping/'
    html = getSource(url)
    images = getImages(html)
    print(images)
```

```python
main()
```

```python
def main():
    import shutil
    url = 'https://infatica.io/blog/web-crawling-vs-web-scraping/'
    html = getSource(url)
    images = getImages(html)
    if images:
        for image in images:
            src = image.get('src')
            if src.startswith('https://infatica.io'):
                r = requests.get(src,stream=True,verify = False)
                if r.status_code == 200:
                    r.raw.decode_content = True
                    #https://infatica.io/blog/web-crawling-vs-web-scraping/        Web-Crawling vs Web-Scraping EXPLAINED
                    filename = src.split('/')[-1]
                    f = open(filename, 'wb')
                    print(filename)
                    shutil.copyfileobj(r.raw,f)
                    f.close()
main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression


df = pd.read_csv('weather.csv')

#print(df.head())

x = df['MinTemp'].values.reshape(-1,1)
y = df['MaxTemp'].values.reshape(-1,1)

#x = df['MinTemp'].values.reshape(-1,1)

#print(x)
plt.scatter(x,y)
plt.ylabel('depdendent')
plt.xlabel('independent')
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
linreg = LinearRegression()

# call the fit function of LinearRegression for training the data
linreg.fit(x_train,y_train)
```

```python
y_pred = linreg.predict(x_test)
```

```python
df1 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})

df1 = df1.head(25)

df1.plot(kind='bar')
```

```python
print(linreg.score(x_test,y_test))
```

```python
# logistic regression

# x - independent variables
# y - dependent ( usually in binary state ie, Male or Female, Yes or No, True or False)



import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# read the csv file
df = pd.read_csv('Student.csv')

print(df.head())
```

```python
# assing columns Self_Study_Daily, Tution_Monthly to x
#  assign column Pass_Or_Fail to y

x = df.drop('Pass_Or_Fail', axis=1)

#print(x.head())

y = df['Pass_Or_Fail']

print(y)
```

```python
# split the data to training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
# creat an instance of Logistic Regression
logreg = LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
df1 = pd.DataFrame({'Actual': y_test,'Predicted': y_pred})

print(df1.head(25))
```

```python
print("score is ",logreg.score(x_test,y_test))
```
